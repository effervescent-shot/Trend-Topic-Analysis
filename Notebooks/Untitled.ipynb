```python
import os
import json
import nltk

import pandas as pd
import numpy as np

rootdir = '../data/'
```

```python
march_01 =  pd.read_csv(rootdir+'2019-03-01_tweetswithlinks_tr.csv', header=None, usecols = np.arange(11),
                        names=['auther_id','created_at','uid','language','text','sname','ttype','RT/Q_uid','RT/Q_sname','Q_text','Q_link'], parse_dates = ['time'],

                        dtype={'tid':float, 'uid':float, 'language':str,'text':str,'sname':str,'ttype':str,'RT/Q_sname':str,'Q_text':str,'Q_link':str  } )
print(len(march_01))
march_01 = march_01[march_01['language'] == 'tr']
march_01.drop(['language'],axis=1, inplace=True)
```
